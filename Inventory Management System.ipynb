# PRODUCTS

## Adding Products

```python
import json
```

```python
try:
    jp=open("record.json","r")
    prod=jp.read()
    jp.close()
    product=json.loads(prod)
except:
    jp=open("record.json","w")
    jp.close()
    product={}

prod_id=input("Enter the Product ID:").upper()
prod_name=input("Enter the Product Name:").upper()
prod_type=input("Enter the Product Type:").upper()
price=int(input("Enter the Product Price:"))
qty=int(input("Enter the Product Quantity:"))

product[prod_id]={'name':prod_name, 'type':prod_type, 'price':price, 'quantity':qty}

j=json.dumps(product)

jp=open("record.json","w")
added=jp.write(j)
jp.close()

if added:
    print("Product Added Successfully!")
```

## Displaying Products

```python
jp=open("record.json","r")
prod=jp.read()
jp.close()

display=json.loads(prod)
count=0
print("\n______________________________________________________")
for i in display:
    count+=1
    print()
    print(count,"\tProduct ID                :",i)
    print("\tProduct Name              :",display[i]['name'])
    print("\tProduct Type              :",display[i]['type'])
    print("\tProduct Price             :",display[i]['price'])
    print("\tProduct Quantity Available:",display[i]['quantity'])
    print("------------------------------------------------------")
```

## Removing Products

```python
jp=open("record.json","r")
prod=jp.read()
jp.close()

remove=json.loads(prod)

pid=input("Enter the Product ID of the product you want to remove:").upper()

found=0
for i in remove:
    if(i==pid):
        found=1
        rm=i
        
if found==0:
    print("Product with ID",pid,"not found")

if found!=0:
    
    del remove[rm]
    
    j=json.dumps(remove)

    jp=open("record.json","w")
    removed=jp.write(j)
    jp.close()
    
    if removed:
        print("\nProduct Removed Successfully!")
    
```

## Search

```python
jp=open("record.json","r")
prod=jp.read()
jp.close()

search=json.loads(prod)

pid=input("Enter the Product ID of the product you want to search:").upper()

found=0
for i in search:
    if(i==pid):
        found=1
        print()
        print("\tProduct ID                :",i)
        print("\tProduct Name              :",search[i]['name'])
        print("\tProduct Type              :",search[i]['type'])
        print("\tProduct Price             :",search[i]['price'])
        print("\tProduct Quantity Available:",search[i]['quantity'])
        print("------------------------------------------------------")
    
if found==0:
    print("Product with ID",pid,"not found!")
```

## Update

```python
jp=open("record.json","r")
prod=jp.read()
jp.close()

update=json.loads(prod)

pid=input("Enter the Product ID of the product you want to update:").upper()

found=0
for i in update:
    if(i==pid):
        found=1
        qty=int(input("Enter the quantity to be increased:"))
        update[pid]["quantity"] += qty

if found==0:
    print("Product with ID",pid,"not found!")
    
if found!=0:
    j=json.dumps(update)

    jp=open("record.json","w")
    updated=jp.write(j)
    jp.close()
    
    if updated:
        print("\nProduct Updated Successfully!")
```

# Sales

## Purchasing products

```python
import time

try:
    js=open("sales.json",'r')
    s=js.read()
    js.close()
    sale=json.loads(s)
except:
    js=open("sales.json",'w')
    js.close()
    sale={}

jp=open("record.json",'r')
prod=jp.read()
jp.close()

product=json.loads(prod)


pid=input("Enter the product ID of the product you want to buy: ").upper()

found=0
for i in product:
    if(i==pid):
        found=1
        name=product[i]['name']
        typ=product[i]['type']
        price=product[i]['price']
        qty=product[i]['quantity']
        if(qty<0 or qty==0):
            print("Product is out of stock")
            break
        print("\n\tProduct Name               : ",name)
        print("\tProduct Type               : ",typ)
        print("\tProduct Price              : ",price)
        print("\tProduct Quantity Available : ",qty)
        print("-------------------------------------------------------------")

if found==0:
        print("Product with ID",pid,"not found!!")
        
if found!=0:
    u_qty=int(input("Enter the quantity you want to buy: "))
    if(u_qty>qty):
        print("Only ",qty," of this product is available.")
    else:
        billing=price*u_qty
        bill_no=time.ctime()
        sale[bill_no]={'PID':pid,'Name':name,'Quantity':u_qty,'Total Amount':billing}
        j=json.dumps(sale)
        js=open("sales.json",'w')
        added=js.write(j)
        js.close()
        if added:
            print("------------------------------------------------------------")
            print("\tBilling ID              : ",bill_no)
            print("\tProduct Name            : ",name)
            print("\tProduct Quantity Bought : ",u_qty)
            print("\tTotal Amount            : ",billing)
            print("------------------------------------------------------------")
            for i in product:
                if(i==pid):
                    product[pid]['quantity']-=u_qty
                    j=json.dumps(product)
                    jp=open("record.json",'w')
                    updated=jp.write(j)
                    jp.close()
                    if not updated:
                        print("Error! Quantity in product file not updated.")
                    


        
```
